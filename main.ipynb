# Visão geral
Leitura do dataset de atrasos de voo.

dataset : umeradnaan/flight-delays-dataset

```python

import pandas as pd
import numpy as np
from kaggle.api.kaggle_api_extended import KaggleApi
import os
```

```python
api = KaggleApi()
api.authenticate()

os.makedirs('data/raw', exist_ok=True)

api.dataset_download_files('umeradnaan/flight-delays-dataset', path='data/raw', unzip=True)  

```

```python

df = pd.read_csv('data//raw//flight_delays.csv', sep=',')
```

# Limpeza de valores ausentes
- Remover linhas com valores nulos: `df.dropna()`
- Remover colunas inteiras se houver nulos: `df.dropna(axis=1)`
- Verificar nulos em `DelayMinutes`: `df["DelayMinutes"].isnull().any()`


```python

df.drop_duplicates(inplace=True)

df['DelayReason'] = df['DelayReason'].fillna('No Delay')
```

# Classificação de atrasos
Criar a coluna `Status`:
- `1` se `DelayMinutes` > 15 (atrasado)
- `0` caso contrário (no horário)

```python
df['Status'] = np.where(df['DelayMinutes'] > 15, 1, 0)
```
